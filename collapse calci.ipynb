```python
import tkinter as tk
from tkinter import ttk

def button_click(number):
    current = entry.get()
    entry.delete(0, tk.END)
    entry.insert(0, current + str(number))

def button_clear():
    entry.delete(0, tk.END)

def button_equal():
    try:
        result = eval(entry.get())
        entry.delete(0, tk.END)
        entry.insert(0, result)
    except:
        entry.delete(0, tk.END)
        entry.insert(0, "Error")

#  conversion functions
def convert_to_cm():
    try:
        value = float(entry.get())
        result = value * 2.54        # inches to centimeters
        entry.delete(0, tk.END)
        entry.insert(0, result)
    except ValueError:
        entry.delete(0, tk.END)
        entry.insert(0, "Invalid Input")

def convert_to_inches():
    try:
        value = float(entry.get())
        result = value / 2.54        # centimeters to inches
        entry.delete(0, tk.END)
        entry.insert(0, result)
    except ValueError:
        entry.delete(0, tk.END)
        entry.insert(0, "Invalid Input")

def convert_to_feet():
    try:
        value = float(entry.get())
        result = value / 30.48       #centimeters to feet
        entry.delete(0, tk.END)
        entry.insert(0, result)
    except ValueError:
        entry.delete(0, tk.END)
        entry.insert(0, "Invalid Input")

def convert_to_meters():
    try:
        value = float(entry.get())
        result = value / 100         #centimeters to meters
        entry.delete(0, tk.END)
        entry.insert(0, result)
    except ValueError:
        entry.delete(0, tk.END)
        entry.insert(0, "Invalid Input")

def convert_to_kg():
    try:
        value = float(entry.get())
        result = value / 2.20462     # pounds to kilograms
        entry.delete(0, tk.END)
        entry.insert(0, result)
    except ValueError:
        entry.delete(0, tk.END)
        entry.insert(0, "Invalid Input")

def convert_to_pounds():
    try:
        value = float(entry.get())
        result = value * 2.20462       # kilograms to pounds
        entry.delete(0, tk.END)
        entry.insert(0, result)
    except ValueError:
        entry.delete(0, tk.END)
        entry.insert(0, "Invalid Input")

def convert_to_fahrenheit():
    try:
        value = float(entry.get())
        result = (value * 9/5) + 32     #Celsius to Fahrenheit
        entry.delete(0, tk.END)
        entry.insert(0, result)
    except ValueError:
        entry.delete(0, tk.END)
        entry.insert(0, "Invalid Input")

def convert_to_celsius():
    try:
        value = float(entry.get())
        result = (value - 32) * 5/9      #Fahrenheit to Celsius
        entry.delete(0, tk.END)
        entry.insert(0, result)
    except ValueError:
        entry.delete(0, tk.END)
        entry.insert(0, "Invalid Input")

#main window code
root = tk.Tk()
root.title("collapsable Calculator")

# entry field code
entry = tk.Entry(root, width=30, borderwidth=5)
entry.grid(row=0, column=0, columnspan=4)

#buttons defining

for (text, row, col) in Buttons:
    button = tk.Button(root, text=text, padx=20, pady=20, command=lambda t=text: button_click(t))
    button.grid(row=row, column=col)

# Clear button
clear_button = tk.Button(root, text="Clear", padx=47, pady=20, command=button_clear)
clear_button.grid(row=5, column=0, columnspan=2)

# Equal button
equal_button = tk.Button(root, text="=", padx=20, pady=20, command=button_equal)
equal_button.grid(row=5, column=2, columnspan=2)

# notebook widget for unit conversions and number pad
notebook = ttk.Notebook(root)
notebook.grid(row=6, column=0, columnspan=4)

# Conversion tabs 
conversion_tabs = [
    ("Length", [convert_to_cm, convert_to_inches, convert_to_feet, convert_to_meters]),
    ("Weight", [convert_to_kg, convert_to_pounds]),
    ("Temperature", [convert_to_fahrenheit, convert_to_celsius]),
   
]

for tab_text, conversion_functions in conversion_tabs:
    tab = tk.Frame(notebook)
    notebook.add(tab, text=tab_text)
    
    row_val = 0
    col_val = 0

    for conversion_func in conversion_functions:
        button = tk.Button(tab, text=conversion_func.__name__.replace("_", " "), padx=10, pady=10, command=conversion_func)
        button.grid(row=row_val, column=col_val)
        col_val += 1
        if col_val > 1:
            col_val = 0
            row_val += 1

# Number pad tab
number_pad_tab = tk.Frame(notebook)
notebook.add(number_pad_tab, text="arithematic operation")

number_pad_buttons = [
    ("7", 1, 0), ("8", 1, 1), ("9", 1, 2), ("/", 1, 3),
    ("4", 2, 0), ("5", 2, 1), ("6", 2, 2), ("*", 2, 3),
    ("1", 3, 0), ("2", 3, 1), ("3", 3, 2), ("-", 3, 3),
    ("0", 4, 0), (".", 4, 1), ("=", 4, 2), ("+", 4, 3),
]

for (text, row, col) in number_pad_buttons:
    button = tk.Button(number_pad_tab, text=text, padx=20, pady=20, command=lambda t=text: button_click(t))
    button.grid(row=row, column=col)

root.mainloop()
```
